Load and inspect data

```python
import pandas as pd
df = pd.read_csv('prices.csv')
df.head()
```

Standardize Column Names

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

Convert Dates to Datetime Format

```python
df['price_dates'] = pd.to_datetime(df['price_dates'], errors='coerce')
```

Check for Missing Values:

```python
df.isnull().sum()
```

Check for Duplicates:

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

Convert Prices to Numeric:

```python
veg_columns = ['bhindi_ladies_finger', 'tomato', 'onion', 'potato', 'brinjal', 'garlic', 'peas', 'methi', 'green_chilli', 'elephant_yam_suran']
df[veg_columns] = df[veg_columns].apply(pd.to_numeric, errors='coerce')
```

Handle Outliers:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for column in veg_columns:
    sns.boxplot(x=df[column])
    plt.show()
```

Export Cleaned Data

```python
df.to_csv('cleaned_vegetable_prices.csv', index=False)
```
